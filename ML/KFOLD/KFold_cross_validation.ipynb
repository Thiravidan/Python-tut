**Importing the datasets**


```python
from sklearn.datasets import load_iris
```

KFold import

```python
from sklearn.model_selection import cross_val_score
```

Import ML models


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
iris = load_iris()
```

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
df.columns
```

```python
iris.target
```

```python
l_scores = cross_val_score(LogisticRegression(max_iter=1000),iris.data,iris.target)
l_scores
```

```python
import numpy as np
np.average(l_scores)
```

```python
d_scores = cross_val_score(DecisionTreeClassifier(),iris.data,iris.target)
d_scores
```

```python
np.average(d_scores)
```

```python
s_scores = cross_val_score(SVC(), iris.data, iris.target)
s_scores
```

```python
np.average(s_scores)
```

```python
r_scores = cross_val_score(RandomForestClassifier(n_estimators=55), iris.data, iris.target)
r_scores
```

```python
np.average(r_scores)
```

```python
print("Score of LogisticRegression =",np.average(l_scores))
print("Score of DecisionTreeClassifier =",np.average(d_scores))
print("Score of SupportVectorMachine =",np.average(s_scores))
print("Score of RandomForestClassifier =",np.average(r_scores))
```

**So Logistic Regression has best score**
